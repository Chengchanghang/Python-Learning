# 非负矩阵分解(NMF)方法及应用
1、**非负矩阵分解**：

- 即Non-negative Matrix Factorization,NMF。在矩阵中所有元素均为非负数约束条件下的矩阵分解方法。

2、**基本原理**：

- 给定一个非负矩阵V，NMF能够找到一个非负矩阵W和一个非负矩阵H，使得矩阵W和H的乘积近似等于V中的值，即：
$$V_{n*m} = W_{n*k} \cdot H_{k*m}$$

- 矩阵$W_{n*k}$：基础图像矩阵，相当于从原矩阵$V_{n*m}$中抽取出来的特征；
- 矩阵$H_{k*m}$：系数矩阵。

3、**NMF优化目标**：

- 最小化W矩阵和H矩阵的乘积与原始矩阵V之间的差别
 - 优化方法一：基于欧氏距离（等式左侧为欧氏距离）
$$argmin\frac{1}{2}||V - W{\cdot}H||^2 = \frac{1}{2}\sum_{i,j}(V_{ij} - W{\cdot}H_{ij})^2$$
 
 - 优化方法二：基于KL散度，损失函数为：
$$argminJ(W, H) = \sum_{i,j}(V_{ij} \ln \frac{V_{ij}}{WH_{ij}} - V_{ij} + WH_{ij})$$

4、**应用领域**

- 图像分析
- 文本挖掘
- 语言处理等领域

# sklearn库及非负矩阵分解
1、**sklearn.decomposition.NMF**：在sklearn.decomposition字库中调用“NMF”即可在Python中实现NMF算法，该方法的**主要参数**有：

- **n_components**：指定分解后矩阵的单个维度k；
- **init**：矩阵W和矩阵H的初始化方式，默认为“nndsvdar”

# 案例：人脸数据特征提取
1、**数据简介**：

- sklearn的子库datasets中提供了Olivetti人脸数据集，该数据集共400个，每个数据时64 * 64大小

2、**研究目的**：

- 由于NMF分解得到的W矩阵相当于从原始矩阵中提取的特征，那么就可以使用NMF对400个人脸数据进行特征提取。

3、**特征提取方法**：

- 通过设定NMF中参数n_components指定分解后矩阵中单个维度k = 6，然后将提取的特征以图像的方式展示出来。

### 1、建立工程，导入sklearn相关包

```python
import matplotlib.pyplot as plt

# 用于进行对比
from sklearn.decomposition import PCA
# 导入NMF算法库
from sklearn.decomposition import NMF

# 导入数据集
from sklearn import datasets

# 导入NumPy库的random子库中的RandomState方法，用于创建随机种子
from numpy.random import RandomState
```

### 2、设置基本参数并加载数据

```python
# 设定图像展示时的排列状况为2行、3列
n_row, n_col = 2, 3

# 设置提取特征的数目
n_components = n_row * n_col

# 设置人脸数据图片的大小
image_shape = (64, 64)

# 加载人脸数据集
# 参数shuffle：若为True，则对数据集重新排序，避免不同分组中有相同人脸图像
# 参数random_state：设定随机种子用于打乱数据
Faces_data = datasets.fetch_olivetti_faces(shuffle = True, random_state = RandomState(0))

faces_data = Faces_data.data

# 400个个体，每个个体中64 * 64 = 4096
type(faces_data), faces_data.shape, faces_data[0]
```

### 3、设置图像的展示方式
1、**matplotlib.pyplot知识补充**：

- **plt.figure(figsize = None)**，该方法用于创建图片
 - 参数：figsize = (宽, 高)，用于指定创建图片的宽和高，指定的宽和高构成元组形式
- **plt.suptitle()**，该方法与**plt.title()**作用类似，都是为图像创建标题，相比较而言使用该方法生成的标题较为美观

```python
def plot_gallery(title, images, n_col = n_col, n_row = n_row):
    
    # 创建图片，指定图片大小（英寸）
    # 参数figsize用于指定新生成图片的代销
    plt.figure(figsize = (2. * n_col, 2.26 * n_row))
    
    # 设置标题及字号大小,suptitle方法用于在图像上方中间增加一个标题
    plt.suptitle(title, size = 16)
    
    for i, comp in enumerate(images):
        
        # 选择画制的子图
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
        
        # 对数值进行归一化，并以灰度图形式显示
        plt.imshow(comp.reshape(image_shape), cmap = plt.cm.gray,
                  interpolation = 'nearest',
                  vmin = -vmax, vmax = vmax)
        
        ## 去除子图的坐标轴标签
        plt.xticks(())
        plt.yticks(())
    
    # 对子图位置及间隔进行调整
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)
```

### 4、创建特征提取的对象NMF，使用PCA作为对比

```python
# 将PCA和NMF降维实例存放在一个列表中
estimators = [
    # PCA降维实例
    ('Eigenfaces - PCA using randomized SVD',
    PCA(n_components = 6, whiten = True)),
    
    # NMF降维实例
    ('Non-negative components - NMF',
    NMF(n_components = 6, init = 'nndsvda', tol = 5e-3))   
]
```

### 5、降维后数据点的可视化

```python
for name, estimator in estimators:
    estimator.fit(faces_data)
    components = estimator.components_
    
    print('%s：%s' % (name, len(components)))
    plot_gallery(name, components[: n_components])

plt.savefig('PCA', dpi = 600)
plt.savefig('NMF', dpi = 600)
plt.show()
```
