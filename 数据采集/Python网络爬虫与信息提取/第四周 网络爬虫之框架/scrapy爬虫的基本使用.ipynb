```python
'''
一、scrapy爬虫的使用步骤：

1、利用命令行（scrapy startproject <工程名>）创建工程和spider模板（scrapy genspider <爬虫名>）

2、编写spider模板

3、编写Item Pipline

4、优化配置策略

二、scrapy爬虫的数据类型

1、Request类  —表示向网络上提交请求的内容

 class scrapy.http.Request()

    Request对象表示一个http请求
    由Spider模块生成，通过Downloader执行
    
    属性或方法       说明
    .url             Request对应的请求url地址
    .method          对应的请求方法，‘get’、‘post’等
    .headers         字典类型风格的请求头         
    .body            请求内容主体，字符串类型
    .meta            用户添加的扩展信息，在Scrapy内部模块间传递信息使用
    .copy()          复制该请求
    
2、response类 —表示从网络上爬取内容的封装类

 class scrapy.http.Response()
     
     Response对象表示一个http响应
     由Downloader生成，通过Spider处理
     
     属性或方法       说明
     .url             Response对应返回信息关联的url地址
     .status          http状态码，默认为200
     .headers         Response对应的头部信息
     .body            Response对应的内容信息，字符串类型
     .flags           Response对应的一组标记
     .request         产生Response类型对应的Request对象
     .copy()          复制该响应

3、item类     —由spider产生的信息封装类

 class scrapy.item.Item()
 
     Item对象表示从HTML页面中提取的信息内容
     由Spider生成，通过Item Piplines处理
     Item类似字典类型，可以按照字典类型操作
     
三、Scrapy爬虫提取信息方法

Scrapy爬虫支持多种HTML信息提取方法

1、BeautifulSoup

2、lxml

3、re

4、XPath Selector

5、CSS Selector

 使用格式：<html>.css('a::attr(href)').extract()，由W3C组织维护并规范
           a    —标签名称 
           href —标签属性
'''
```
