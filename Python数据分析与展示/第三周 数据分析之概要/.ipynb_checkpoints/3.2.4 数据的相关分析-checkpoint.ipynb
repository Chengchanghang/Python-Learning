```python
'''
数据的相关分析

   相关分析：（两个变量X、Y）
       
       X增大，Y增大，则两个变量正相关；
       X增大，Y减小，则两个变量负相关；
       X增大，Y无变化，则两个变量不相关。
    
   相关性度量：
   
       协方差：              ∑[Xi - E（X）][Yi - E（Y）]
                 cov(X, Y) = ----------------------------   (i = 1, 2, ..., n) 
                                       n - 1
                 若协方差大于零，则X、Y存在正相关；
                 若协方差小于零，则X、Y存在负相关；
                 若协方差等于零，则二者独立无关
                 （协方差对于相关性检验不精确）
       
       Pearson相关系数：
                                     ∑[Xi - E(X)][Yi - E(Y)]
                        r = ----------------------------------------  (i = 1, 2, ..., n)
                             √∑[Xi - E(X)]^2 * √∑[Yi - E(Y)]^2
                -1 =< r =< 1
                r的绝对值在0.8-1.0之间为极强相关；0.6-0.8之间为强相关；0.4-0.6中等强度相关；0.2-0.4弱相关；0-0.2极弱相关或不相关         

   适用于Series和DataFrame类型的相关分析函数
   
       函数                    说明
       .cov()                  计算协方差矩阵
       .corr()                 计算相关系数矩阵，Pearson、Spearman、Kendall等系数
'''     
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
price = pd.Series([3.84, 22.93, 12.75, 22.6, 12.33], index = ['2008', '2009', '2010', '2011', '2012'])
m2 = pd.Series([8.18, 18.38, 9.13, 7.82, 6.69], index = ['2008', '2009', '2010', '2011', '2012'])

price.corr(m2)
```

```python
price.cov(m2)
```

```python
plt.plot(price, color = 'r')
plt.plot(m2, color = 'b')
plt.xlabel('\n年份\n', fontproperties = 'SimHei', fontsize = 17)
plt.ylabel('\nM2\n', fontsize = 17)
plt.title('\n房屋增幅与M2增幅的相关性\n', fontproperties = 'SimHei', fontsize = 20)
plt.xlim(2007, 2013)
plt.grid(True)
plt.show()
```
