```python
'''
pyplot的中文显示

由于pyplot并不支持中文显示，故可通过以下两种方法实现图形中中文的显示：

  1、通过rcParams修改字体实现中文显示，它是Matplotlib库用于改变全局字体的相关资源库
  
     注意：
         
         使用该方法会改变图形中出现的所有字符，不仅包括中文字符，还包括横纵坐标轴上显示的数值
  
     rcParams的属性
     
         属性                 说明
         font.family          用于显示字体的名字
         ---------------------------------------
            中文字体          说明             
            SimHei            中文黑体
            Kaiti             中文楷体
            LiSu              中文隶书
            FangSong          中文仿宋
            YouYuan           中文幼圆
            STSong            华文宋体
         ---------------------------------------
         font.style           字体风格，正常‘normal’或斜体‘italic’
         font.size            字体大小，整数字号或‘large’、‘x-small’
  
  2、在有中文输入的地方增加一个属性：fontproperties
  
     注意：
     
         该方法的使用并不会改变全局字体（即所有出现的字体），而是针对具体想要改变的字体进行修改
         
         建议使用第二种中文显示方法，可以避免修改全局字体，而只针对需要修改的字体进行修改。
'''
```

```python
# 通过rcParams实现中文显示

import matplotlib.pyplot as plt
import matplotlib

# 为使中文能够显示，使用matplotlib中的rcParams将font.family修改为中文字体（此处修改为SimHei黑体）
matplotlib.rcParams['font.family'] = 'SimHei'
plt.plot([3, 1, 4, 5, 2], 'g')
plt.ylabel('纵轴（值）')
plt.savefig('test2', dpi = 600)
plt.show()
```

```python
# 使用Matplotlib库中的rcParams方法显示中文 —— 修改了全局字体

import matplotlib.pyplot as plt
import matplotlib
import numpy as np

# 设置字体为中文宋体，通过该方法指定字体名称图形中所有字体均会修改为此类字体（包括横纵坐标轴的数值）
matplotlib.rcParams['font.family'] = 'STSong' 

# 设置字体大小为20，通过该方法指定字体大小会改变图形中所有字体的大小（包括横纵坐标的数值）
matplotlib.rcParams['font.size'] = 20 

a = np.arange(0.0, 5.0, 0.02) # 生成0-5且间隔为0.02的一维数组

# 设置横纵纵轴标签
plt.xlabel('横轴，时间')
plt.ylabel('纵轴，振幅')

# 绘制余弦曲线，曲线为红色、虚线
plt.plot(a, np.cos(2*np.pi*a), 'r--')
plt.show()
```

```python
# 使用fontporperties属性显示中文 —— 修改指定字体

import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0.0, 5.0, 0.02)

# 通过properties属性显示中文，但并不改变其他位置的字符
plt.xlabel('横轴，时间', fontproperties = 'SimHei', fontsize = 20)
plt.ylabel('纵轴，振幅', fontproperties = 'SimHei', fontsize = 20)

plt.plot(a, np.cos(2 * np.pi * a), 'g--')
plt.show()
```
