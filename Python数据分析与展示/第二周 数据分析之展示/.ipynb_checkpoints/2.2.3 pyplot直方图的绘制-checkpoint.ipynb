```python
'''
pyplot直方图的绘制

    直方图可用于观察一组数据的的分布情况。

    plt.hist(x, bins, normed)
    
    关键参数意义：
    
        x   ：数组或数组序列
        bins：直方图的个数，即分为多少个柱状
        normed：当normed = 1时，即将每个直方中出现元素的个数归一化为元素出现概率
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mu, sigma = 100, 20 # 均值和标准差
a = np.random.normal(loc = mu, scale = sigma, size = 100)
print(max(a), min(a))
plt.hist(a, bins = 10, normed = 1, histtype = 'stepfilled', facecolor = 'r', alpha = 0.75)

plt.title('直方图', fontproperties = 'SimHei', fontsize = 30)
plt.show()
```
