```python
'''
pyplot子绘图区域

1、简单（整齐的绘图区域）

   plt.sunplot(nrows, ncols, plot_number)表示将绘图区域分割为nrows * ncols个子区域，当前绘制的图形位于plot_number子区域

   注意：
       
       该函数是将绘图区域按照指定参数整齐分割为多个子绘图区域
       
2、复杂绘图区域

   2.1、plt.subplot2grid(GridSpec, CurSpec, colspan = 1, rowspan = 1)
   
       该函数可作为subplot的辅助函数来划分复杂的绘图区域。

       基本理念：设定规则的网格，选择其中的网格，根据选中的网格确定占用的行列区域数量，编号从0开始

       例：plt.subplot2grid((3 ,3), (1, 0), colsapn = 2)，即为ax2子区域

           (3, 3)同subplot中参数nrows和ncols，表示将绘图区域分割为3行3列共9个子绘图区域，行列子区域编号均从0开始
           (1, 0)表示分割的子区域中的行编号为1，列编号为0的子区域
           colspan表示在列方向的延伸，即横向延伸，colspan = 2即将在行编号为1处等同于ax4的子区域横向延伸为ax2
           rowspan表示在行方向的延伸，即纵向延伸

           plt.subplot2grid((3, 3), (0, 0), colspan = 3)，即为ax1子区域
           plt.subplot2grid((3, 3), (1, 2), rowspan = 2)，即为ax3子区域
           plt.subplot2grid((3, 3), (2, 0))，即为ax4子区域
           plt.subplot2grid((3, 3), (2, 1))，即为ax5子区域

                                                  ncols

                                     0           1             2 

                       0                           ax1                    

                               ----------------------------------------
             nrows                                     |              
                       1                  ax2          |              
                                                       |              
                               ------------------------|     ax3      
                                          |            |              
                       2            ax4   |     ax5    |              
                                          |            |    
        注意：

            该方法相较于subplot函数而言，可以产生较为复杂的子区域，但每次调用subplot2grid函数均需对绘图区域的分割进行约束

   2.2、使用GridSpec类与subplot函数实现同样的功能
   
        GridSpec类 — Matplotlib库中用来区域设计和选定的类
        
        引入方式：import matplotlib.GridSpec as gridspec 
        
        gs = gridspec.GridSpec(3, 3) 在该区域中设计一个3行3列共9个子区域的网格
        
        ax1 = plt.subplot(gs[0, :]) gs[0, :]中逗号前边表示行编号，逗号后边表示列编号，“：”表示三列网格全部包括
        ax2 = plt.subplot(gs[1, :-1]) 此处也是不包括终止值，即-1表示的子区域
        ax3 = plt.subplot(gs[1:, 2])
        ax4 = plt.subplot(gs[2, 0])
        ax5 = plt.subplot(gs[2, 1])
        
        注意：
        
            使用该方法无需每次对绘图区域进行划分，只需通过GridSpec划分一次即可
'''
```
