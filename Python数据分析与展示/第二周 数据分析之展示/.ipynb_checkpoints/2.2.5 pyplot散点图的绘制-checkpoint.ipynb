```python
'''
pyplot散点图的绘制

    plt.scatter() 此处不使用plt.scatter函数绘制散点图
    
    采用面向对象的方法，此时绘制图形不再是对应的函数（如：散点图对应plt.scatter），而是对象对应的方法。
'''
```

```python
# 使用面向对象的方法绘制散点图

import numpy as np
import matplotlib.pyplot as plt

# subplots函数用于创建一个图和多个子图，参数为空时，即等价于plt.subplots(111)
# 此处令subplots生成的图表和对应的子区域作为fig和ax对象
fig, ax = plt.subplots()
ax.plot(10 * np.random.randn(100), 10 * np.random.randn(100), 'o')
ax.set_title('简单散点图', fontproperties = 'SimHei', fontsize = 30)

plt.show()
```
