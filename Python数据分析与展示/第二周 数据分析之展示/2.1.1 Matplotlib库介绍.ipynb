```python
'''
Matplotlib库

1、Matplotlib库是Python中优秀的数据可视化第三方库，由各种可视化类构成，内部结构复杂。

2、Matplotlib库的使用（受到MATLAB启发）

   为不顾及内部复杂结构方便使用Matplotlib库，该库提供Matplotlib.pyplot子库（pyplot是绘制各类可视化图形的命令子库，即仅调用pyplot子库就可以使用Matplotlib库中所有可视化类，相当于快捷方式）
   
   import matplotlib.pyplot as plt（导入pyplot子库）
   
3、pyplot的绘图区域

   plt.subplot(nrows, ncols, plot_number)
   
   subplot函数将绘图区域分割为以nrows作为横轴数量、ncols作为纵轴数量的子区域
   
   subplot函数能够通过前两个参数在全局绘图区域中创建一个分区体系，并通过最后一个参数定位到一个子绘图区域。
   
   例：plt.subplot(3,2,4)（等价于subplot(324)），即表示将绘图区域分为三行两列（即将绘图区域分为六个子区域），当前的图形绘制区域位于子区域4。
                                         ncols
                                           |
                    子区域1                |                子区域2
         ----------------------------------|-------------------------------------
                                           |
                                           |
                                           |
  nrows             子区域3                |                子区域4
                                           |
                                           |
                                           |
         ----------------------------------|-------------------------------------
                    子区域5                |                子区域5
                                           |
'''
```

```python
# plot函数输入一个列表的情形

import matplotlib.pyplot as plt

# plot函数若输入参数为一个列表，则该参数被当做y轴来处理，x轴即为该列表元素对应的索引值
plt.plot([3,1,4,5,2]) 
plt.ylabel('grade') # 设置y轴标签

# 利用savefig()函数将输出图形存储，保存为PNG文件，第一个参数为文件名，第二个参数dpi可修改输出质量（dpi指在每一英寸空间中包含点的数量）
plt.savefig('test', dpi = 600) 
plt.show() # 显示图形
```

```python
# plot函数中输入两个列表的情形

import matplotlib.pyplot as plt

# plot函数中输入两个列表时，则第一个列表表示x轴，第二个列表表示y轴
plt.plot([1,2,3,4], [2,4,6,8])
plt.ylabel('grade')
plt.xlabel('ID')

# axis用于设置x、y轴的尺度，即表示x轴起始于1，终止于5；y轴起始于0，终止于9
plt.axis([1, 5, 0, 9])
plt.savefig('test1', dpi = 600)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 该函数为能量衰减函数
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

a = np.arange(0.0, 5.0, 0.02)

plt.subplot(211) # 将绘图区域分为两部分，并将当前图形绘制在子区域1
plt.plot(a, f(a)) # 绘制能量衰减函数

plt.subplot(2,1,2) # 将当前图形绘制在子区域2
plt.plot(a, np.cos(2 * np.pi * a), 'r--') # 绘制余弦函数，并通过参数“r--”表示该余弦曲线用虚线表示
plt.show()
```
