```python
'''
pyplot文本显示方法

    pyplot的文本显示函数
    
          常用函数                    说明
        plt.xlabel()            对x轴增加文本标签
        plt.ylable()            对y轴增加文本标签
        plt.title()             对图形整体增加文本标签（即图形标题，位于图形正上方）
        plt.text()              在图形任意位置增加文本
        plt.annotate()          在图形中增加带箭头的注解（annotate）
        
        plt.annotate(s, xy = arrow_crd, xytext = text_crd, arrowprops = dict)
        
            参数意义：
                
                s：表示待注释的字符串
                xy：表示箭头所在位置
                xytext：表示文本显示的位置
                arrowprops：字典类型，定义整个箭头显示的属性
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成0.0-5.0且间隔为0.02的一维数组
a = np.arange(0.0, 5.0, 0.02)

# 作正弦曲线
plt.plot(a, np.cos(2 * np.pi * a), 'y-.')

# 设置x、y轴的标签，并利用properties属性实现中文显示，color指定标签的字体颜色
plt.xlabel('横轴，时间', fontproperties = 'SimHei', fontsize = 15, color = 'green')
plt.ylabel('纵轴，振幅', fontproperties = 'SimHei', fontsize = 15, color = 'red')

# 添加正弦曲线的标题，使用美元符号括起来的这部分为LaTeX格式的文本
plt.title(r'正弦曲线 $y = cos(2\pi x)$', fontproperties = 'SimHei', fontsize = 25, color = 'blue')

# 在图形中(2, 1)位置显示指定文本，其中2表示x轴坐标，1表示y轴坐标
# plt.text(2, 1, r'$\mu = 100$', fontsize = 15)

# 第一个参数即为待注释的文本，第二个参数表示箭头所在位置坐标点，第三个参数为文本所在位置坐标点
# 第四个参数定义箭头相关属性，其中第一个表示箭头颜色，第二个表示箭头起始端到结束端按0.1比例缩进（与指向点和文本之间留有一定空白），第三个表示箭头宽度
plt.annotate(r'$\mu = 100$', xy = (2, 1), xytext = (3, 1.5), 
            arrowprops = dict(facecolor = 'red', shrink = 0.1, width = 2))

# 设定坐标轴，x轴为-1～6，y轴为-2～2
plt.axis([-1, 6, -2, 2])

# grid用于设定图形中是否显示网格，若为True则显示网格；反之不显示
plt.grid(True)

# 将图形保存为PNG格式图片
plt.savefig('test3', dpi = 600)
plt.show()
```
