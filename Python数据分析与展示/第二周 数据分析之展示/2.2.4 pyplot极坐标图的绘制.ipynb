```python
'''
极坐标图

    即中心点为原点，按照360度的方向来绘制坐标形成的图形即为极坐标图
    
    该图形一般不常用。
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 20 # 表示绘制极坐标图的数据个数

# 生成20个0.0-2 * np.pi且不包括终止值的随机数
theta = np.linspace(0.0, 2 * np.pi, N, endpoint = False)

# 生成20个0-1之间服从均匀分布的随机数
radii = 10 * np.random.rand(N)

width = np.pi / 4 * np.random.rand(N)

# projection表示绘制何种图形，polar即为极坐标图，并将其赋值给ax
# 此处绘制极坐标图使用面向对象方式（通过不同对象之间的调用实现指定功能）
ax = plt.subplot(111, projection = 'polar')

# left表示绘制极坐标图中颜色区域是从哪个位置开始的
# height表示从中心点向边缘绘制的长度
# width表示每一个子区域对应的面积
bars = ax.bar(left = theta, height = radii, width = width, bottom = 0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.9) # 参数数值越小，颜色越淡
    
plt.show()
```
